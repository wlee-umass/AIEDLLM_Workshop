```python
q4H1m = list(dfq4['"q4H1m"'])
q4H1m = [s.replace('"', "") for s in q4H1m]
q4H1m = list(filter(None, q4H1m))
q4H1m_str = ' '.join(q4H1m)

q4H1t = list(dfq4['"q4H1t"'])
q4H1t = [s.replace('"', "") for s in q4H1t]
q4H1t = list(filter(None, q4H1t))
q4H1t_str = ' '.join(q4H1t)

q4H2m = list(dfq4['"q4H2m"'])
q4H2m = [s.replace('"', "") for s in q4H2m]
q4H2m = list(filter(None, q4H2m))
q4H2m_str = ' '.join(q4H2m)

q4H2t = list(dfq4['"q4H2t"'])
q4H2t = [s.replace('"', "") for s in q4H2t]
q4H2t = list(filter(None, q4H2t))
q4H2t_str = ' '.join(q4H2t)

q4HintComments = q4H1m_str + q4H1t_str + q4H2m_str + q4H2t_str
```

## Put Q1 and Q4 Hm comments together as a list of strings

```python
"""q1H1m_str,
    q1H2m_str,
    q1H3m_str,
    q4H1m_str,
    q4H2m_str"""

user_feedback = [
    q1HintComments,
    q4HintComments
]
```

```python
# For fine-tuning:

# model = DistilBertForMaskedLM.from_pretrained('distilbert-base-uncased')
# tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

## Use the BART model, tokenize the comments, and pass them to BART for summarization

```python
model_name = 'facebook/bart-large-cnn'
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
```

```python
inputs = tokenizer(user_feedback, return_tensors="pt", truncation=True, padding=True, max_length=512)
```

```python
summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=60, early_stopping=True)
```

## Decoding phase

```python
summaries = [tokenizer.decode(ids, skip_special_tokens=True, clean_up_tokenization_spaces=False) for ids in summary_ids]
```

```python
for summary in summaries:
    print(summary)
```
